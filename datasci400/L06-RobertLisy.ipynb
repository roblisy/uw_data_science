```python
# use a library called BeautifulSoup for web scraping.
import bs4
import requests
```

```python
url = "https://data.fivethirtyeight.com/"

page = requests.get(url).text
r = requests.get(url)
html_content = r.text

soup = bs4.BeautifulSoup(html_content, "html.parser")

# Set an integer to count the number of links on this page
num_links = 0
# loop over the soup object and find all the links.
# example from here (https://www.crummy.com/software/BeautifulSoup/bs4/doc/)
for link in soup.find_all("a"):
    # if you want to see the links, uncomment the line below
    # print(link.get('href'))
    num_links += 1
    
print('The number of links on this page is ' + str(num_links))
```
